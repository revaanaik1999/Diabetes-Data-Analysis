```python
from sklearn.feature_selection import f_classif
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Split the data into input features and target variable
X = df[finalCols]
y = df['readmitted']

# Perform ANOVA test
f, pval = f_classif(X, y)

# Create a dataframe of feature scores
anova_df = pd.DataFrame({'Feature': finalCols, 'F-Score': f, 'P-Value': pval})

# Print the dataframe sorted by F-Score
print(anova_df.sort_values(by=['F-Score'], ascending=False))
```

## Outlier detection using K-means

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from scipy.spatial.distance import cdist
import matplotlib.pyplot as plt
import numpy as np
import warnings

warnings.filterwarnings('ignore')

# Extract features
X = df[selected_features]

# Scale numerical features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Perform k-means clustering
kmeans = KMeans(n_clusters=5, random_state=42)
kmeans.fit(X_scaled)

# Get cluster labels
labels = kmeans.labels_

# Perform PCA on the scaled data for visualization
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Plot scatter plot of PCA components
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=labels, cmap='rainbow')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.title('K-means Clustering Results')

# Define legend colors
colors = ['red', 'blue', 'green', 'orange', 'purple']

# Add legends
legend_elements = [plt.Line2D([0], [0], marker='o', color='w', label='Cluster {}'.format(i), 
                              markerfacecolor=colors[i], markersize=10) for i in range(len(set(labels)))]
plt.legend(handles=legend_elements, loc='best')

# Get cluster centers
centers = kmeans.cluster_centers_

# Calculate the distances between each point and its cluster center
distances = cdist(X_scaled, centers)

# Get the minimum distance for each point
min_distances = np.min(distances, axis=1)

# Define the threshold for marking outliers
threshold = np.percentile(min_distances, 95)

# Mark the outliers on the scatter plot
outliers = np.where(min_distances > threshold)[0]
plt.scatter(X_pca[outliers, 0], X_pca[outliers, 1], marker='x', color='black')

plt.show()
```
